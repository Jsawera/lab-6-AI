```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset (same as homeprices.csv)
data = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "price": [550000, 565000, 610000, 680000, 725000]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Display dataset
print("Dataset:")
print(df)

# Plotting the data
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.title('Area vs Price')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (USD)')
plt.show()

# Preparing data for Linear Regression
X = df[['area']]  # Independent variable
y = df['price']   # Dependent variable

# Creating and training the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predicting prices for areas in the dataset
dataset_predictions = model.predict(X)
print("\nPredicted Prices for Dataset Areas:")
for area, price in zip(df['area'], dataset_predictions):
    print(f"Area: {area} sq ft, Predicted Price: ${price:.2f}")

# Plotting regression line
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.plot(df['area'], dataset_predictions, color='blue', linewidth=2)
plt.title('Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (USD)')
plt.show()
```

```python
# Importing necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Dataset
data = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "price": [550000, 565000, 610000, 680000, 725000]
}

# Preparing data for Linear Regression
X = np.array(data["area"]).reshape(-1, 1)  # Independent variable
y = np.array(data["price"])                # Dependent variable

# Creating and training the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Custom areas for prediction
custom_areas = np.array([[5000], [8000], [9000]])

# Predicting prices for custom areas
predicted_prices = model.predict(custom_areas)
print("\nPredicted Prices for Custom Areas:")
for area, price in zip(custom_areas.flatten(), predicted_prices):
    print(f"Predicted price for a home with area {area} sq ft: ${price:.2f}")
```

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
data = {
    "age": [1, 2, 3, 4, 5],
    "price": [22000, 21000, 20000, 18000, 17000]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Display dataset
print("Dataset:")
print(df)

# Plotting the data
plt.scatter(df['age'], df['price'], color='blue', marker='o')
plt.title('Car Age vs Resale Price')
plt.xlabel('Age (years)')
plt.ylabel('Resale Price (USD)')
plt.show()

# Preparing data for Linear Regression
X = df[['age']]  # Independent variable (2D array required for sklearn)
y = df['price']  # Dependent variable

# Creating and training the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predicting prices for custom car ages
custom_ages = [[6], [8], [10]]  # Predicting for cars with 6, 8, and 10 years age
predicted_prices = model.predict(custom_ages)

print("\nPredicted Resale Prices for Custom Car Ages:")
for age, price in zip([6, 8, 10], predicted_prices):
    print(f"Predicted price for a car with age {age} years: ${price:.2f}")

# Plot regression line
plt.scatter(df['age'], df['price'], color='blue', marker='o', label='Actual Data')
plt.plot(df['age'], model.predict(X), color='red', label='Regression Line')
plt.scatter([6, 8, 10], predicted_prices, color='green', marker='x', label='Predicted Points')
plt.title('Regression Line with Predictions')
plt.xlabel('Age (years)')
plt.ylabel('Resale Price (USD)')
plt.legend()
plt.show()
```
